# Doctor Visit Analysis using Python

**Importing necessary libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"/content/DoctorVisits.csv")
```

```python
df.head()
```

```python
df=df.drop(columns="Unnamed: 0",axis=1)
```

**Visualize and analyse the maximum,minimum and median income**

```python
y=list(df.income)
plt.boxplot(y)
plt.show()
```

**Find out the no of days reduced activity of male and female seperatly due to illness**

```python
df.groupby(['gender','reduced']).mean()
```

**Visualize is there is any missing values in the dataset based on a heat map**

```python
sns.heatmap(df.isnull(),cbar=False,cmap='viridis')
```

**Find out the correlation between variables in the given dataset correlation between different variables**

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),cbar=True,annot=True,cmap='Blues')
```

**Analyse how the income of a patient affects the no of visits to the hospital**

```python
plt.figure(figsize=(10,10))
plt.scatter(x='income',y='visits',data=df)
plt.xlabel('income')
plt.ylabel('visits')
```


 **Count and visualize the number of males and females affected by illness**

```python
sns.histplot(df.gender,bins=2)
```

**Plot a horizontal bar chart to analyze the reduced days of activity due to illness based on gender**

```python
label=['yes','no']
y=df[df['freepoor']=='yes']
n=df[df['freepoor']=='no']
x=[y.shape[0],n.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("% of people getting govt health insurance due to low income")
plt.show()
label=['yes','no']
y=df[df['private']=='yes']
n=df[df['private']=='no']
x=[y.shape[0],n.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("% of people getting private health insurance")

plt.show()
label=['yes','no']
y=df[df['freerepat']=='yes']
n=df[df['freerepat']=='no']
x=[y.shape[0],n.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("% of people getting govt health insurance due to old age")
plt.show()
```

**Plot a Horizontal bar chart to analyze the reduced days of activity due to illness based on gender** **bold text**

```python
db=df.groupby('gender')['reduced'].sum().to_frame().reset_index()
plt.barh(db['gender'],db['reduced'],color=['cornflowerblue','lightseagreen'])
plt.title('Bar Chart')
plt.xlabel('gender')
plt.ylabel('reduced activity')
plt.show()
```
